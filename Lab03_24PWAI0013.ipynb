```python
square = lambda x: x ** 2
print(f"The square of 5 is: {square(5)}")
print(f"The square of 9 is: {square(9)}")
print(f"The square of 12 is: {square(12)}")
```

## Task 9

```python
numbers = [1, 2, 3, 4, 5]
# map() applies the lambda to every item in the list
squares = list(map(lambda x: x ** 2, numbers))
print(f"Original numbers: {numbers}")
print(f"Squared numbers:  {squares}")
```

## Task 10

```python
marks = [45, 67, 82, 30, 90, 55]
# filter() checks each number — keeps it only if greater than 50
above_50 = list(filter(lambda x: x > 50, marks))
print(f"All marks:        {marks}")
print(f"Marks above 50:   {above_50}")
```
