# AI Password Strength Checker

```python
attempts = 5
used_attempts = 0
print(f"Welcome to AI Password Strength Checker Lite!")
print(f"You have {attempts} attempts to create a strong password.\n")
while used_attempts < attempts:
    password = input("Enter your password: ")
    # Conditions
    has_length = len(password) >= 8
    has_digit = any(char.isdigit() for char in password)
    has_upper = any(char.isupper() for char in password)
    # Count conditions met
    conditions_met = 0
    if has_length:
        conditions_met += 1
    if has_digit:
        conditions_met += 1
    if has_upper:
        conditions_met += 1
    # Strength check
    if conditions_met == 0:
        strength = "Very Weak"
    elif conditions_met == 1:
        strength = "Weak"
    elif conditions_met == 2:
        strength = "Moderate"
    else:
        strength = "Strong"
    used_attempts += 1
    print(f"Password Strength: {strength}")
    # Winning condition
    if strength == "Strong":
        print(f"\nYou created a strong password!")
        print(f"AI Training Level: Beginner → Intermediate")
        break
    # Remaining attempts
    remaining = attempts - used_attempts
    print(f"Attempts left: {remaining}\n")
# Losing condition
if strength != "Strong":
    print(f"\nFinal Result: Password is not strong enough.")
```
