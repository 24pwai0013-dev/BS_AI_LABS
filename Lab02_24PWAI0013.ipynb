## Task 2

```python
# Ask the user for two numbers, using float() to allow decimals
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
# Ask the user which operation they want to perform
operator = input("Enter operator (+, -, *, /, **, %): ")
# Check if the operator entered is one of the valid/allowed options
if operator not in ["+", "-", "*", "/", "**", "%"]:
    # If operator is invalid, print error and stop
    print("Invalid operator!")
# Check for division by zero before dividing
elif operator == "/" and num2 == 0:
    print("Cannot divide by zero!")
# Perform the correct calculation based on the operator chosen
elif operator == "+":
    result = num1 + num2
    print(f"{num1} + {num2} = {result}")
elif operator == "-":
    result = num1 - num2
    print(f"{num1} - {num2} = {result}")
elif operator == "*":
    result = num1 * num2
    print(f"{num1} * {num2} = {result}")
elif operator == "/":
    result = num1 / num2
    # Show 2 decimal places for division results
    print(f"{num1} / {num2} = {result:.2f}")
elif operator == "**":
    result = num1 ** num2
    # Show 2 decimal places for power results
    print(f"{num1} ** {num2} = {result:.2f}")
elif operator == "%":
    result = num1 % num2
    print(f"{num1} % {num2} = {result}")
```

## Task 3

```python
# Ask the user for a starting and ending number
start = int(input("Enter starting number: "))
end   = int(input("Enter ending number:   "))
print("\n--- Number Check ---")
# Use a while loop to go through every number in the range
current = start
while current <= end:
    # Skip the number entirely if it is divisible by 7 (continue jumps to next iteration)
    if current % 7 == 0:
        current += 1
        continue
    # Check both conditions together first (even AND multiple of 5)
    if current % 2 == 0 and current % 5 == 0:
        print(f"{current} → Even & Multiple of 5!!")
    # Check if only even
    elif current % 2 == 0:
        print(f"{current} → Even")
    # Check if only a multiple of 5
    elif current % 5 == 0:
        print(f"{current} → Multiple of 5!")
    # Odd and no special property
    else:
        print(f"{current} → Odd")
    # Move to the next number
    current += 1
# TASK-03 | Challenge B – Password Strength Checker
print("\n--- Password Strength Checker ---")
# Ask the user to enter a password
password = input("Enter a password to check: ")
# Check length first — must be at least 6 characters
if len(password) < 6:
    print("Too short!")
else:
    # Nested check: look for at least one digit inside the password
    has_digit   = any(char.isdigit()   for char in password)
    # Nested check: look for at least one uppercase letter
    has_upper   = any(char.isupper()   for char in password)
    if not has_digit:
        print("Add at least one number")
    elif not has_upper:
        print("Add at least one capital letter")
    # Only reaches here if length OK + has digit + has uppercase
    else:
        print("Strong password – good choice!")
```
